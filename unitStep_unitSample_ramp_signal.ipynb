```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = np.arange(-5, 6 ,1)
n
```

```python
unit_sample = np.zeros(len(n))
unit_sample[n == 0] = 1
unit_sample
print(unit_sample)

plt.stem(n, unit_sample)
plt.grid(True)
```

```python
unit_step = np.where(n >= 0 , 1, 0)
plt.stem(n, unit_step)
plt.grid(True)
```

```python
ramp = np.where(n >= 0, n, 0)
plt.stem(n, ramp)
plt.grid(True)
```

```python
plt.Figure(figsize=(100, 100))

plt.subplot(5, 1, 1)
plt.title("Unit Sample")
plt.stem(n, unit_sample)
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.grid(True)

plt.subplot(5, 1, 3)
plt.title("Unit Step")
plt.stem(n, unit_step)
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.grid(True)

plt.subplot(5, 1, 5)
plt.title("Ramp")
plt.stem(n, ramp)
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.grid(True)
```
