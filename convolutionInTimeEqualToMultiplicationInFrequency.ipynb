```python
import numpy as np
```

```python
def DFT(x):
    N = len(x)
    X = np.zeros(N, dtype=complex)
    
    for k in range(N):
        for n in range(N):
            X[k] += x[n] * np.exp(-1j * 2 * np.pi * k * n / N)
    
    return X
```

```python
def convolution(x, h):
    N = len(x)
    M = len(h)
    
    y = np.zeros(N + M - 1)
    
    for n in range(N + M - 1):
        for k in range(N):
            if 0 <= n - k < M:
                y[n] += x[k] * h[n - k]
    
    return y
```

```python
def zero_pad(x, size):
    padded = np.zeros(size)
    padded[:len(x)] = x
    return padded
```

```python
x = np.array([1, 2, 3])
h = np.array([1, 1, 1])
```

```python
y = convolution(x, h)
```

```python
# Length for DFT
L = len(y)

# Zero padding
x_pad = zero_pad(x, L)
h_pad = zero_pad(h, L)
```

```python
X = DFT(x_pad)
H = DFT(h_pad)
Y_freq = X * H
```

```python
Y_time = DFT(y)
```

```python
print("DFT of convolution (Y_time):")
print(np.round(Y_time, 4))

print("\nProduct of DFTs (Y_freq):")
print(np.round(Y_freq, 4))
```
