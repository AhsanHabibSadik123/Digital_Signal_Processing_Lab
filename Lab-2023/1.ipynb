```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def signal(t, f):
    return t, (np.sin(2 * np.pi * f * t))
```

```python
f1 = 30
t1 = np.linspace(0, 1, 1000)
t1, x1 = signal(t1, f1)
plt.plot(t1, x1)
```

```python
sample_rate = 100
sampling_time = 1 / sample_rate
ts1 = np.arange(0, 1, sampling_time)
ts1, xs1 = signal(ts1, f1)
plt.stem(ts1, xs1)
```

```python
def DFT(x):
    N = len(x)
    X = np.zeros(N, dtype=complex)
    for k in range(N):
        for n in range(N):
            X[k] += x[n] * np.exp(-1j * 2 * np.pi * k * n / N)
    return X
```

```python
X1 = DFT(xs1)
```

```python
def plot_spectrum(X):
    magnitude = np.abs(X)
    phase = np.angle(X)
    k = np.arange(len(X))
    
    plt.Figure(figsize=(10, 8))
    
    plt.subplot(3, 1, 1)
    plt.stem(k , magnitude)
    plt.title("Magnitude")
    plt.xlabel("k")
    plt.ylabel("X[k]")
    
    plt.subplot(3, 1, 3)
    plt.stem(k , phase)
    plt.title("Phase")
    plt.xlabel("k")
    plt.ylabel("Phase(Radian)")
```

```python
plot_spectrum(X1)
```

```python
f2 = 60
t2 = np.linspace(0, 1, 1000)
t2, x2 = signal(t2, f2)
plt.plot(t2, x2)
```

```python
sample_rate = 100
sampling_time = 1 / sample_rate
ts2 = np.arange(0, 1, sampling_time)
ts2, xs2 = signal(ts2, f2)
plt.stem(ts2, xs2)
```

```python
X2 = DFT(xs2)
```

```python
plot_spectrum(X1)
```

```python
plot_spectrum(X2)
```
